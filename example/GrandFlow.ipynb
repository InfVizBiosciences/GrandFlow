```python
# 写入fastq文件
from grandflow.io import FastqWriter

with FastqWriter("output.fq.gz") as writer:
    writer.writeRecord("dog", "GATTACA", [35]*7)
    writer.writeRecord("cat", "CATTACA", [35]*7)
```

```python
# %load read_vcf.py

from grandflow.io import VcfReader


vr = VcfReader('data/test.vcf')
for item in vr:
    print('chrom:' + item.chrom)
    print('id:' + item.id)
    print(item.info)
    print('SVTYPE:' + item.info['SVTYPE'])
    break
```

## 生成 Pipeline

```python
from grandflow.Pipeline import Pipeline, Task

path = './'

pipeline_name = Pipeline('pipeline_test', path)

sjm_config_dict = {

  'memory': '16G',
  'thread': 8,
  'queue': 'all.q'

}

# 新建任务
# 参数分别是
# 任务名称
# 任务 cmd， 可能会包含 {}
# cmd_paras 命令行的参数
# sjm_paras sjm的参数,如果仅生成shell，可以不输入
task1 = Task(
    'task1',
    'echo {task_name}',
    {'task_name':'task1'},
    sjm_paras= sjm_config_dict)


pipeline_name.add_task(task1)


# task2 多任务并行

task2 = Task(
    'task1',
    'echo {task_name}',
    {'task_name': ['task2_0', 'task2_1', 'task2_2','task2_3']},
    sjm_paras= sjm_config_dict)


pipeline_name.add_task(task2, prev_task=task1)
```

```python
print(pipeline_name.sjm())
```
